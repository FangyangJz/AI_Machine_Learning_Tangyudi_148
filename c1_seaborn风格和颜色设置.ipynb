```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*.5) * (7-i) * flip)
```

```python
sinplot()
```

```python
sns.set()  # 用sns默认模板设置
sinplot()
```

# 一、风格

#### sns提供了5种风格
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6)/2
sns.boxplot(data=data)
```

```python
data.shape
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

#### 去掉上和右侧边框线

```python
sinplot()
sns.despine()
```

```python
# 破开原点坐标轴连接
sns.violinplot(data=data)
sns.despine(offset=40)
```

```python
# despine 指定保留的边框
sns.boxplot(data=data, palette='deep')
sns.despine(left=True)  # True 为消失
```

#### with 开辟一个独立的块，制定风格

```python
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### 设置整体风格, 设置不同图大小

```python
sns.set_context('paper')
plt.figure(figsize=(6,3))
sinplot()
```

```python
sns.set_context('poster')
plt.figure(figsize=(6,3))
sinplot()
```

```python
sns.set_context('notebook', font_scale=3.5, rc={'lines.linewidth':5.5})
plt.figure(figsize=(6,3))
sinplot()
```

# 二、调色板
* color_palette() 能传入任何matplotlib支持的颜色
* color_palette() 不写参数，默认颜色
* set_palette() 设置所有图的颜色

```python
# 分类色板
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

#### 6个默认的颜色循环主题：deep, muted, pastel, bright, dark, colorblind

```python
# hls 为默认的颜色空间
sns.palplot(sns.color_palette('hls', n_colors=10))
```

```python
sns.palplot(sns.color_palette(n_colors=10))
```

```python
sns.set_context('poster')
sns.boxplot(data=data, palette=sns.color_palette('hls', 8))
```

```python
# sns.hls_palette() 可以用来调整亮度和饱和度
sns.palplot(sns.hls_palette(n_colors=8, l=0.7, s=0.8))
```

```python
# 成对的颜色
sns.palplot(sns.color_palette('Paired', 8))
```

```python
sns.boxplot(data=data, palette=sns.color_palette('Paired', 8))
```

### (离散指定颜色)使用xkcd颜色来命名颜色
xkcd包含了一套众包努力的针对随机RGB色的命名， 产生了954个可以随时通过xkcd_rgb字典中调用的命名颜色

```python
plt.plot([0,1], [0,1], sns.xkcd_rgb['pale aqua'], lw=3)
plt.plot([0,1], [0,2], sns.xkcd_rgb['medium green'], lw=3)
plt.plot([0,1], [0,3], sns.xkcd_rgb['denim blue'], lw=3)
```
